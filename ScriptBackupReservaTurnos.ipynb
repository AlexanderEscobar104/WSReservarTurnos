# [dbo].[udf_turnos]

```sql
/****** Object:  UserDefinedFunction [dbo].[udf_turnos]    Script Date: 10/04/2023 8:47:02 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- create a multi-statement table-valued udf para retornar los turnos de acuerdo a la fecha, hora y tiempo proporcionado 
CREATE function [dbo].[udf_turnos](
@ID_SERVICIO AS BIGINT,
@FECHA_INICIO DATETIME,
@FECHA_FIN DATETIME,
@HORA_APERTURA AS TIME,
@HORA_CIERRE AS TIME,
@DURACION AS INT
)

--tabla de retono  
returns @turnos TABLE (
        ID_SERVICIO BIGINT,
        FECHA_TURNO DATETIME,
        HORA_INICIO TIME,
        HORA_FIN TIME,
		ESTADO varchar(30)
    )
as


begin


DECLARE @HORA_INICIO TIME
DECLARE @HORA_FIN TIME
DECLARE @ESTADO varchar(30)

-- agregamos la hora del servicio a las fechas 
SET @FECHA_INICIO = DATEADD(MINUTE,-@DURACION, CONVERT(DATETIME,  @FECHA_INICIO) + ' ' +  CONVERT(DATETIME, @HORA_APERTURA))
SET @FECHA_FIN = DATEADD(MINUTE,-@DURACION, CONVERT(DATETIME,  @FECHA_FIN) + ' ' +  CONVERT(DATETIME, @HORA_CIERRE))

		-- realiza el recorrido de las dos fechas 
		WHILE (@FECHA_INICIO < @FECHA_FIN)
		BEGIN

		 --cambia el estado si la hora o fecha ya paso 
		  IF @FECHA_INICIO < DATEADD(MINUTE,-@DURACION, GETDATE())
		  BEGIN
			SET @ESTADO = 'INACTIVO'
		  END 
		  ELSE
		  BEGIN
			SET @ESTADO = 'ACTIVO'
		  END 
	  
			-- agregamos la duracion 
			SELECT @FECHA_INICIO = DATEADD(MINUTE,@DURACION, @FECHA_INICIO) 
			SET @HORA_FIN =  DATEADD(MINUTE,@DURACION, @FECHA_INICIO) 

			--insertamos en la tambla de retorno  
			INSERT INTO @turnos (id_servicio
					   ,fecha_turno
					   ,hora_inicio
					   ,hora_fin
					   ,estado)
				  VALUES
					   (@ID_SERVICIO
					   ,@FECHA_INICIO
					   ,CONVERT(TIME, @FECHA_INICIO, 108)
					   ,@HORA_FIN
					   ,@ESTADO)

		END

 
   return;
 
end;
GO
```

# [dbo].[Insert_Turnos]

```sql
/****** Object:  StoredProcedure [dbo].[Insert_Turnos]    Script Date: 10/04/2023 8:47:02 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- =============================================
-- Author:		AMESCOBAR
-- Create date: 04-ABR-2023
-- Description:	procedimiento para crear turnos
-- =============================================
CREATE PROCEDURE [dbo].[Insert_Turnos] 
	-- parameters
	
	@Fecha_inicio DATE = null, 
	@Fecha_fin DATE = null, 
	@Id_Servicio INT = 0

AS

DECLARE @HORA_APERTURA AS TIME
DECLARE @HORA_CIERRE AS TIME
DECLARE @DURACION AS INT

DECLARE @TEMP_TURNOS AS TABLE 
		(ID_SERVICIO BIGINT,
        FECHA_TURNO DATETIME,
        HORA_INICIO TIME,
        HORA_FIN TIME,
		ESTADO varchar(50))

BEGIN

	-- valida que no existan turnos en la fecha indicada
	IF NOT EXISTS(SELECT * FROM TURNOS T WHERE T.FECHA_TURNO = @Fecha_inicio)
	BEGIN
	
		--VALIDA SI EL SERVICIO EXISTE 
		IF EXISTS(SELECT * FROM SERVICIOS S WHERE S.ID_SERVICIO = @Id_Servicio)
		BEGIN
		 --OBTIENE LOS DATOS DE LA TABLA SERVICIOS 
		 SELECT @HORA_APERTURA = S.HORA_APERTURA,@HORA_CIERRE = S.HORA_CIERRE ,@DURACION = S.DURACION FROM SERVICIOS S WHERE S.ID_SERVICIO = @Id_Servicio
	 
		 --ELIMINO TODOS LOS DATOS PARA MANTENER LOS ULTIMOS REGISTROS
		--DELETE FROM turnos

 
		 --CONSULTA LA UDF, INSERTA LOS DATOS EN TABLA TEMPORAL 
		 INSERT INTO @TEMP_TURNOS
		 SELECT * FROM  dbo.udf_turnos(@Id_Servicio,@Fecha_inicio,@Fecha_fin, @HORA_APERTURA, @HORA_CIERRE, @DURACION) 

		 --INSERTA LOS DATOS EN LA TABLA PRODUCTIVA
		 INSERT INTO turnos
		 SELECT * FROM @TEMP_TURNOS
	
		--OBTIENE LOS DATOS INSERTADOS
		 SELECT * FROM @TEMP_TURNOS

		END

	END
	ELSE
	BEGIN
		RAISERROR('Ya Se Genero Turnos Para La Fecha Indicada.',16,1);
	    RETURN;
	END;



END
GO
```

# [dbo].[Select_Turnos]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Turnos]    Script Date: 10/04/2023 8:47:02 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- =============================================
-- Author:		AMESCOBAR
-- Create date: 04-ABR-2023
-- Description:	procedimiento para seleccionar turnos
-- =============================================
CREATE PROCEDURE [dbo].[Select_Turnos] 
	-- parameters
	

AS

BEGIN

		--ACTUALIZAMOS LOS TURNOS YA INSCATIVOS POR FECHA Y HORA		
		UPDATE TURNOS SET ESTADO = 'INACTIVO'  WHERE CONVERT(DATETIME,  FECHA_TURNO) + ' ' +  CONVERT(DATETIME, HORA_INICIO)  <  CONVERT(DATETIME, GETDATE())
		--TRAE TODOS LOS TURNOS 	
		SELECT * FROM  TURNOS T  ORDER BY FECHA_TURNO DESC

END
GO
```
